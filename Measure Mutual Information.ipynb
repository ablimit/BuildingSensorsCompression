```python
def histogram(S,_mean,_std,k=2,l=10):
    """ Compute the histogram for a sequence and identify peaks
        S=a pandas series containing the data
        _mean,_std - the mean and standard deviation of the sequence
        k - the range of the histogram is from mean-k*std to mean+k*std with two extra bins
            for counting the extremes
        l - we use l bins per std, for a total of 2kl bins
    """
    S=sorted(S.dropna())
    B=arange(_mean-k*_std,_mean+k*_std,_std/l)
    counts=np.zeros(len(B)+1)
    i1=0; i2=0
    for j in range(len(B)):
        while (S[i2]<B[j]) & (i2<len(S)-1): 
            i2+=1
        counts[j]=i2-i1
        i1=i2
    counts[-1]=len(S)-i1
    return B,counts,S
```

```python
def Compute_Distortion(S,reps):
    _total = 0.0
    for s in S:
        D=np.min([(r-s)**2 for r in reps])
        _total+=D
    return _total/len(S)

def find_reps(S,B,counts):
    # compute mid-point of each bin
    midpoints=[(B[i]+B[i+1])/2.0 for i in range(len(B)-1)]
    RDF=pd.DataFrame({'counts':counts[1:-1],'midpoint':midpoints})
    RDF=RDF.sort(columns=['counts'],ascending=False)

    reps = set([])
    error = 1e20 
    for i in range(shape(RDF)[0]):
        counts,midpoint=RDF.iloc[i]
        #print counts, midpoint
        treps=reps.copy()
        treps.add(midpoint)
        D=Compute_Distortion(S,treps)

        if (error-D)>_std/10:
            #print i,'midpoint=',midpoint,'error=',D,'improvement=',error-D #,reps
            reps=treps.copy()
            error=D
        #else:
            #print 'skipping',i, 'midpoint=',midpoint
    return error,reps   
```

```python
from collections import Counter
def CommonValues(S,top_n=3,P_covered=0.95):
    C=Counter(S)
    total=float(len(S))
    count=0.0
    Answer=[]
    for e in C.most_common(top_n):
        count += e[1]
        Answer.append(e)
        #print e,count,count/total
        if (count/total>P_covered):
                break
    fractionExplained = count/total
    return fractionExplained,Answer
```

```python
def plot_a_day(A,ax1,graph_title='',with_legend=False):
    
    p1,=ax1.plot(A.index,A['Zone_Temperature'],'b',label='Zone_Temperature',linewidth=2.0);
    p3,=ax1.plot(A.index,A['Actual_Cooling_Setpoint'],'y',label='Actual_Cooling_Setpoint',linewidth=2.0);
    p4,=ax1.plot(A.index,A['Actual_Heating_Setpoint'],'c',label='Actual_Heating_Setpoint',linewidth=2.0);
    p5,=ax1.plot(A.index,A['Temperature'],'b--',label='Temperature',linewidth=2.0);
    B=A['Heating_Command'][A['Heating_Command']>0]
    p6,=ax1.plot(B.index,B,'k',label='Heating_Command',linewidth=2.0);
    p7,=ax1.plot(A.index,A['Occupied_Command']*5+70,'g--',label='Occupied_Command',linewidth=2.0)

    plt.ylabel('Degrees Fahrenheit',fontsize=20);
    ax2 = ax1.twinx()
    p2,=ax2.plot(A.index,A['Actual_Supply_Flow'],'r',label='Actual_Supply_Flow',linewidth=2.0);
    plt.ylabel('Cubic Feet per Minute',fontsize=20);
    grid()
    ax1.set_title(graph_title)
    if with_legend:
        #ax1.cla(); ax2.cla()
        ax1.legend(loc=2)
        ax2.legend(loc=3)
        #[p1, p2,p3,p4,p5,p6], ["Zone Temperature", "Supply Flow", "Cooling Set Point","Heating Set Point","External Temperature"],"upper right");
```

```python
from datetime import timedelta
from time import strftime
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

room='1145'
DateRange=daterange(datetime.date(2013, 12, 9), datetime.date(2013, 12, 16))

DF=A[room]
DF=DF.iloc[32:,:] # Remove undefined rows
DF['date']=DF.index.date

subplots_W=3
subplots_H=3
fig, ax = plt.subplots(subplots_H,subplots_W)
i=0        
for single_date in DateRange:
    date_string=strftime("%a %Y-%m-%d", single_date.timetuple())   
    DayData=DF[DF['date']==single_date]
    plot_a_day(DayData,ax[i/subplots_W,i%subplots_W],graph_title=date_string)
    i+=1
plot_a_day(DayData,ax[i/subplots_W,i%subplots_W],graph_title=date_string, with_legend=True)
```
