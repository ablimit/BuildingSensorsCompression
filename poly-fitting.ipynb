```python
from numpy import *
import pylab

# data to fit
x = arange(0,10,1)
y = x**2+100*random.rand(10)

# fit the data with a 4th degree polynomial
z4 = polyfit(x, y, 0) 
p4 = poly1d(z4) # construct the polynomial 

z5 = polyfit(x, y, 2)
p5 = poly1d(z5)

xx = linspace(0, 10, 100)
pylab.plot(x, y, 'o', xx, p4(xx),'-g', xx, p5(xx),'-b')
pylab.legend(['data to fit', '1st degree poly', '2nd degree poly'])
#pylab.axis([0,1,0,1])
pylab.show()
```
